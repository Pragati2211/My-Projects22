#### 1.3. Dataset Features : <a name="Datasetfeatures"></a>


Partitioning Dataset into the Set of Descriptive Features and the Target Feature
Next, we partition diamonds df columns into the set of descriptive features and the target.


```python
# get the list of descriptive features
descriptive_features = df[['cut', 'depth', 'color', 'price']] # Accessing descriptive data
prediction_features = df['carat'] # Accessing target variable

# prediction data is given, importing the data
prediction_data = {'cut': 'Good', 'depth': 60, 'color': 'D', 'price': 'premium'}

# Appending the data in dataframe:
final_descriptive_features= descriptive_features.append(prediction_data, ignore_index = True) 
```

### Q1 Part A
#### 2. One Hot Encoding

A one hot encoding is a representation of categorical variables as binary vectors.

This first requires that the categorical values be mapped to integer values.

Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.



```python
# for other categorical features (with > 2 levels), 
# One hot encoding

# get the list of categorical descriptive features
nom_data = final_descriptive_features.columns[final_descriptive_features.dtypes==object].tolist()

# if a categorical descriptive feature has only 2 levels, define only one binary variable
for i in nom_data:
    n = len(final_descriptive_features[i].unique())
    if (n == 2):
        final_descriptive_features[i] = pd.get_dummies(final_descriptive_features[i], drop_first=True)

# for other categorical features (with > 2 levels), use regular one-hot-encoding, if a feature is numeric, it will be untouched
final_descriptive_features = pd.get_dummies(final_descriptive_features)

# use regular one-hot-encoding 
# features, depth and carat are numeric, it will remain same 
Data = pd.get_dummies(final_descriptive_features)
# displaying last 10 rows of hot encoded data:

Data.tail(10)
```

All three catagorical descriptive features are converted into binary using one hot encoding

##### K-NN Regressor with Scikit-learn
Scikit-learn is a Machine Learning Python library that features a variety of classification, regression, and clustering algorithms.

To demonstrate how to perform KNN with Python we will use the diamonds dataset where we will use all of the quanitative features to predict the price of the diamond. So, go ahead and call in the diamonds dataset again and remove all the qualitative features, and place the target variable, price, at the end of the dataframe.

```python
from sklearn import preprocessing
#for k=1
from sklearn.neighbors import KNeighborsClassifier
# The ".values" part below converts the data frame to a 2-dimensional numpy array
# The ".values" part below converts the data frame to a 2-dimensional numpy array
X_train = new_data.iloc[:212,:9] # Using entire dataset for training
y_train = new_data.iloc[:212,9:] 

y_test = new_data.iloc[212,:9]
y_test = pd.DataFrame(y_test).T

```

```python

# k=1 carat prediction using KNNRegressor()

one_knn_regressor = KNeighborsRegressor(n_neighbors=1, p=2)
one_knn_regressor.fit(X_train, y_train)

one_knn_prediction = one_knn_regressor.predict(y_test)
print(f"1-KNN Prediction using KNNRegressor(): {one_knn_prediction}")
```

```python
#for k=5
from sklearn.neighbors import KNeighborsClassifier

# K=5  carat prediction using KNNRegressor()

five_knn_regressor = KNeighborsRegressor(n_neighbors=5, p=2)
five_knn_regressor.fit(X_train, y_train)

five_knn_prediction = five_knn_regressor.predict(y_test)
print(f"1-KNN Prediction using KNNRegressor(): {five_knn_prediction}")
```

```python
# for k=10
from sklearn.neighbors import KNeighborsClassifier
# k=10 carat prediction using KNNRegressor()

ten_knn_regressor = KNeighborsRegressor(n_neighbors=10, p=2)
ten_knn_regressor.fit(X_train, y_train)

ten_knn_prediction = ten_knn_regressor.predict(y_test)
print(f"1-KNN Prediction using KNNRegressor(): {ten_knn_prediction}")
```
